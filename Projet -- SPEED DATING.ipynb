```python
#Categorisation des matches:
matches = data.match.apply(lambda x: "yes" if x==1
                                      else "no"
                                      )
matches
```

```python
#Sum des matches :
pie_chart_matches = matches.value_counts()
pie_chart_matches
```

```python
#Catégorisation des decisions prises lors des dates:
decision = data.dec.apply(lambda x: "yes" if x ==1 
                                     else "no"
                                     )
decision
```

```python
#Sum des decisions :
pie_chart_decision = decision.value_counts()
pie_chart_decision
```

```python
#Catégorisation des decisions prises par les partenaires lors des dates:
decision_o = data.dec_o.apply(lambda x: "yes" if x ==1 
                                     else "no")
decision_o
```

```python
#Sum des decisions prises par les partenaires:
pie_chart_decision_o = decision_o.value_counts()
pie_chart_decision_o
```

```python
explode = [0.0,0.2]
```

```python
# Création de trois graphiques via plt.subplot()
plt.figure(figsize=(10,10))

## Création du premier pie chart
plt.subplot(221)
plt.pie(pie_chart_decision, explode=explode,labels=pie_chart_decision.index, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title("Decision at night of event")
plt.legend(pie_chart_decision, loc='upper left', bbox_to_anchor=(0,1))

## Création d'une seonde pie chart
plt.subplot(222)
plt.pie(pie_chart_decision_o, explode=explode,labels=pie_chart_decision_o.index, autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.title("Decision of partner at night of event") 
plt.legend(pie_chart_decision_o, loc='upper left', bbox_to_anchor=(1,1)) 

## Création d'une troisième pie chart
plt.subplot(223)
plt.pie(pie_chart_matches, explode=explode,labels=pie_chart_matches.index, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title("matches at night of event")
plt.legend(pie_chart_decision, loc='upper left', bbox_to_anchor=(0,1))

plt.show()
```
