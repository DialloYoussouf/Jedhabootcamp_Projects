#### Rating

```python
rating_cleaned = rating.select(
  F.col("userID"),
  F.col("placeID"),
  F.when(F.col("rating") == "2", 1).otherwise(0).alias("label")
)
```

```python
display(
  rating_cleaned.groupBy("placeID")
        .pivot("label")
        .count())
```
