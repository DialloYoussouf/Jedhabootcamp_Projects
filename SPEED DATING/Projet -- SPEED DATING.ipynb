```python
#Categorisation des matches:
matches = data.match.apply(lambda x: "yes" if x==1
                                      else "no" )
matches
```

```python
#Somme des match :
matches_count = matches.value_counts()
matches_count
```

```python
#Somme des match en % :
matches_count_percent = np.round((matches_count /(matches_count.sum()))*100,2)
matches_count_percent
```

```python
# Visualisation des "match" en %
fig, ax = plt.subplots(figsize = (8,5))
ax.bar( matches_count_percent.index, matches_count_percent.values )
plt.title(" Pourentage des match lors des dates")
plt.show()
```

```python
# Visualisation de l'importance des "like" en fonction des "attr" de la part des partenaires dans la prise des decisions:
# like_o : affection de la partenaire 
# attr_o : attirance par la partenaire
# dec_o (0 == "no" & 1 == "yes") : decision de la partenaire des des dates 
sns.relplot(data = data, x= "attr_o", y = "like_o", hue="dec_o"  )
plt.show()
```

==> On remaqrue qu'Ã  partir d'une notation de "like_o" et "attr_o" >= 6, la prise de decision ("dec_o") Ã  tendance Ã  Ãªtre "yes" chez la partenaire.

```python
# Visualisation de l'importance des "like" en fonction des "attr" de la part des partenaires dans la prise des decisions:
# like:   affection  
# attr : l'attirance de la partenaire
# dec (0 == "no" & 1 == "yes") :  decision lors des dates
sns.relplot(data = data, x= "attr", y = "like", hue = "dec")
plt.show()
```

==> Nous avons aussi une forte corrÃ©lation entre les notations de "like" et "attr" >= 6, donc la prise de decision ("dec") aura tendance Ã  Ãªtre "yes".

On obtient un "match" en cas de decision favorable ("yes") de la part des 2 partenaires.
Par consÃ©quent,  nous pouvons supposer que ces variables sont importantes dans l'obtention d'une seconde date. 

En effet, nous partons du principe que l'obtention d'un "match" est synonyme d'une seconde date.

```python
# CaracÃ©risation de l'importance de la rÃ©ligion lors des dates
imprace = data.imprace.apply(lambda x: "Absolument important(score=10)" if x==10
                                else "ExtrÃªmement important(score=9)" if x ==9
                                else "TrÃ¨s TrÃ¨s important(score=8)" if x==8
                                else "TrÃ¨s important(score=7)" if x ==7
                                else "Important(score=6)" if x==6
                                else "moyen(score=5)" if x==5
                                else "TrÃ¨s peu important(score=4)" if x==4
                                else "Pas assez important(score=3)" if x==3
                                else "Pas important(score=2)" if x==2
                                else "Pas du tout important(score=1)" if x ==1
                                else "Totalement indiffÃ©rent(score=0)")
imprace
```

```python
imprace_count = imprace.value_counts().sort_values()
imprace_count
```

```python
# Visualisation de l'importance de la rÃ©ligion lors des dates
fig, ax = plt.subplots(figsize = (10,6))
ax.barh(imprace_count.index, imprace_count.values)
plt.title("L'importance d'avoir la mÃªme rÃ©ligion chez les 2 partenaires")
plt.xlabel("Somme des notes attribuÃ©es par les participants")
plt.show()
```
